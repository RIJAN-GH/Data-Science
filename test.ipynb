### if-else Statements

```python
age = int(input("Enter Your age:"))
if age >= 18:
    print("You are eligible for citizenship.")
else:
    print("You are not eligible.")
```

```python
num = int(input("Enter any number:"))
if num < 0:
    print(f"The number({num}) is negative number.")
elif 0 <= num <= 50:
    print("The number is positive and fall under 50.")
else:
    print("The number is greater than 50.")
```

### Break and Continue
---

```python
for i in range(10):
    if i == 5:
        break
    print(i, end = " ")
print()
for i in range(10):
    if i == 5:
        continue
    print(i, end = " ")
```

### While loop
---

```python
n = int(input("Enter n number:"))
sum = 0
i = 0
while i<=n:
    sum = sum + i
    i += 1
print(sum)
```

### Function
---

```python
def prime(num):
    if num == 0 or num == 1:
        return 1
    for i in range(2, num):
        if (num % i) == 0:
            return 0
        else:
            return 1

n = int(input("Enter any number:"))
result = prime(n)
if result == 1:
    print("The num is prime.")
else:
    print("The number is not prime.")

```

### Visualization

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

### Line Chart

```python

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

### Bar PLot

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")     # add a title
plt.ylabel("# of Academy Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./titanic.csv")
df.head(5)
```

```python
# countplot
sns.countplot(data=df, x='Sex')
df['Survived'].value_counts()
```

```python
# Bivariate Analysis
sns.barplot(x='Pclass',y='Age',hue='Sex',data=df)
```

### Pie Chart

```python
 # pie chart

df['Sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

### Box Plot

```python
# Bivarient Analysis


#Box plot (Numerical - Categorical)

sns.boxplot(x='Sex',y='Age',hue='Survived',data=df)
```

### Histogram

```python

# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['Age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.displot(df['Age'], kind="hist", kde=True)
```

```python
# Bivariant analysis
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Died')
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived')

plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
